#Lecture3

```python
url = requests.get("https://coinmarketcap.com/")
```

```python
doc = BeautifulSoup(url.text, "html.parser")
```

```python
print(doc.prettify())
```

```python
tbody = doc.tbody
tbody
```

```python
trs = tbody.contents
trs
```

```python
prices = {}

for tr in trs[:30]:
	name, price = tr.contents[2:4]
	fixed_name = name.p.string if name.p else name.get_text(strip=True)
	fixed_price = price.a.string if price.a else price.get_text(strip=True)

	prices[fixed_name] = fixed_price

print(prices)
```
